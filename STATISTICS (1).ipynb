  Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance.

````python
class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
        if len(outcomes) != len(probabilities):
            raise ValueError("Lengths of outcomes and probabilities must match.")
        
        self.outcomes = outcomes
        self.probabilities = probabilities
    
    def expected_value(self):
        return sum(outcome * probability for outcome, probability in zip(self.outcomes, self.probabilities))
    
    def variance(self):
        mean = self.expected_value()
        squared_deviations = [(outcome - mean) ** 2 for outcome in self.outcomes]
        return sum(probability * squared_deviation for probability, squared_deviation in zip(self.probabilities, squared_deviations))
````

````python
# Define outcomes and probabilities
outcomes = [1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.4]

# Create a DiscreteRandomVariable object
random_variable = DiscreteRandomVariable(outcomes, probabilities)

# Calculate expected value and variance
expected_val = random_variable.expected_value()
variance_val = random_variable.variance()

print("Expected Value:", expected_val)
print("Variance:", variance_val)
````

 Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance.

````python
import numpy as np

def generate_samples(distribution, size, *params):
    """
    Generate random samples from a given probability distribution and calculate mean and variance.

    Args:
    - distribution: A string indicating the probability distribution (e.g., 'binomial', 'poisson').
    - size: The number of samples to generate.
    - *params: Additional parameters required for the specified distribution (e.g., n and p for binomial, lam for Poisson).

    Returns:
    A tuple containing the generated samples, mean, and variance.
    """
    if distribution == 'binomial':
        n, p = params
        samples = np.random.binomial(n, p, size)
    elif distribution == 'poisson':
        lam = params[0]
        samples = np.random.poisson(lam, size)
    else:
        raise ValueError("Unsupported distribution. Supported distributions are 'binomial' and 'poisson'.")

    mean = np.mean(samples)
    variance = np.var(samples)
    return samples, mean, variance

# Example usage:
distribution = 'binomial'
n = 10
p = 0.5
size = 1000

samples, mean, variance = generate_samples(distribution, size, n, p)

print("Generated Samples:", samples[:10], "...")
print("Mean:", mean)
print("Variance:", variance)
````

 Write a Python function to calculate the probability density function (PDF) of a continuous random
variable for a given normal distribution.


````python
import numpy as np

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution at a given point x.

    Args:
    - x: The point at which to evaluate the PDF.
    - mean: The mean (mu) of the normal distribution.
    - std_dev: The standard deviation (sigma) of the normal distribution.

    Returns:
    The PDF value at the given point x.
    """
    coefficient = 1 / (std_dev * np.sqrt(2 * np.pi))
    exponent = -((x - mean) ** 2) / (2 * std_dev ** 2)
    pdf_value = coefficient * np.exp(exponent)
    return pdf_value

# Example usage:
x_value = 1.5
mean_value = 0
std_dev_value = 1

pdf_result = normal_pdf(x_value, mean_value, std_dev_value)
print("PDF at x =", x_value, ":", pdf_result)
````

Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.

````python
import numpy as np

def exponential_cdf(x, rate):
    """
    Calculate the cumulative distribution function (CDF) of the exponential distribution at a given point x.

    Args:
    - x: The point at which to evaluate the CDF.
    - rate: The rate parameter (lambda) of the exponential distribution.

    Returns:
    The CDF value at the given point x.
    """
    cdf_value = 1 - np.exp(-rate * x)
    return cdf_value

# Example usage:
x_value = 2
rate_value = 0.5

cdf_result = exponential_cdf(x_value, rate_value)
print("CDF at x =", x_value, ":", cdf_result)
````

 Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

````python
import math

def poisson_pmf(k, rate):
    """
    Calculate the probability mass function (PMF) of the Poisson distribution at a given point k.

    Args:
    - k: The point at which to evaluate the PMF.
    - rate: The average rate (lambda) of the Poisson distribution.

    Returns:
    The PMF value at the given point k.
    """
    pmf_value = (math.exp(-rate) * rate**k) / math.factorial(k)
    return pmf_value

# Example usage:
k_value = 2
rate_value = 3

pmf_result = poisson_pmf(k_value, rate_value)
print("PMF at k =", k_value, ":", pmf_result)
````

Apply z-test to find which layout is successful.

````python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Number of successes (purchases) and total trials (visitors) for old and new layouts
successes_old = np.sum(old_layout)
trials_old = len(old_layout)
successes_new = np.sum(new_layout)
trials_new = len(new_layout)

# Perform z-test for two proportions
z_stat, p_value = proportions_ztest([successes_old, successes_new], [trials_old, trials_new])

# Interpret the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("There is a significant difference between the old and new layouts.")
    if z_stat < 0:
        print("The new layout leads to a higher conversion rate.")
    else:
        print("The old layout leads to a higher conversion rate.")
else:
    print("There is no significant difference between the old and new layouts.")
````

 A tutoring service claims that its program improves students' exam scores. A sample of students who

Use the below code to generate samples of respective arrays of marks:

```python

before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])

after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

```

Use z-test to find if the claims made by tutor are true or false.



````python
import numpy as np
from scipy.stats import norm

# Data
before_program = np.array([75, 80, 85, 70, 90, 78, 92, 88, 82, 87])
after_program = np.array([80, 85, 90, 80, 92, 80, 95, 90, 85, 88])

# Calculate the difference in scores
score_difference = after_program - before_program

# Calculate the mean and standard deviation of the score differences
mean_diff = np.mean(score_difference)
std_dev_diff = np.std(score_difference, ddof=1)  # ddof=1 for sample standard deviation

# Number of samples
n = len(score_difference)

# Calculate the standard error of the mean difference
sem_diff = std_dev_diff / np.sqrt(n)

# Perform z-test
z_stat = mean_diff / sem_diff

# Calculate p-value using the cumulative distribution function of the standard normal distribution
p_value = 2 * (1 - norm.cdf(abs(z_stat)))  # multiplied by 2 for a two-tailed test

# Interpret the results
alpha = 0.05  # significance level
if p_value < alpha:
    print("The tutoring program has a significant impact on students' exam scores.")
    if z_stat < 0:
        print("The scores have decreased after the program.")
    else:
        print("The scores have increased after the program.")
else:
    print("There is no significant impact of the tutoring program on students' exam scores.")
````

A company is testing two different website layouts to see which one leads to higher click-through rates
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
freedom, and p-value.


Use the following data:

```python

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

````python
import numpy as np
from scipy.stats import t

def ab_test(layout_a_clicks, layout_b_clicks, alpha=0.05):
    # Calculate sample statistics
    mean_a = np.mean(layout_a_clicks)
    mean_b = np.mean(layout_b_clicks)
    std_dev_a = np.std(layout_a_clicks, ddof=1)  # ddof=1 for sample standard deviation
    std_dev_b = np.std(layout_b_clicks, ddof=1)
    n_a = len(layout_a_clicks)
    n_b = len(layout_b_clicks)

    # Calculate pooled standard deviation
    pooled_std_dev = np.sqrt(((n_a - 1) * std_dev_a ** 2 + (n_b - 1) * std_dev_b ** 2) / (n_a + n_b - 2))

    # Calculate t-statistic
    t_stat = (mean_a - mean_b) / (pooled_std_dev * np.sqrt(1/n_a + 1/n_b))

    # Calculate degrees of freedom
    degrees_of_freedom = n_a + n_b - 2

    # Calculate p-value using the cumulative distribution function of the t-distribution
    p_value = 2 * (1 - t.cdf(abs(t_stat), degrees_of_freedom))  # multiplied by 2 for a two-tailed test

    # Interpret the results
    if p_value < alpha:
        print("The difference in click-through rates between the two layouts is statistically significant.")
        if t_stat < 0:
            print("Layout B leads to a higher click-through rate.")
        else:
            print("Layout A leads to a higher click-through rate.")
    else:
        print("There is no significant difference in click-through rates between the two layouts.")

    return t_stat, degrees_of_freedom, p_value

# Example usage with provided data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

t_statistic, degrees_of_freedom, p_value = ab_test(layout_a_clicks, layout_b_clicks)
print("t-statistic:", t_statistic)
print("Degrees of freedom:", degrees_of_freedom)
print("p-value:", p_value)
````

````python
import numpy as np
from scipy.stats import t

def paired_t_test(pre_intervention_scores, post_intervention_scores, alpha=0.05):
    # Calculate the differences between pre- and post-intervention scores
    differences = np.array(post_intervention_scores) - np.array(pre_intervention_scores)

    # Calculate sample statistics
    mean_diff = np.mean(differences)
    std_dev_diff = np.std(differences, ddof=1)  # ddof=1 for sample standard deviation
    n = len(differences)

    # Calculate standard error of the mean difference
    sem_diff = std_dev_diff / np.sqrt(n)

    # Calculate t-statistic
    t_stat = mean_diff / sem_diff

    # Calculate degrees of freedom
    degrees_of_freedom = n - 1

    # Calculate p-value using the cumulative distribution function of the t-distribution
    p_value = 2 * (1 - t.cdf(abs(t_stat), degrees_of_freedom))  # multiplied by 2 for a two-tailed test

    # Interpret the results
    if p_value < alpha:
        print("The intervention had a significant impact on math scores.")
        if t_stat < 0:
            print("The math scores decreased after the intervention.")
        else:
            print("The math scores increased after the intervention.")
    else:
        print("There is no significant impact of the intervention on math scores.")

    return t_stat, degrees_of_freedom, p_value

# Example usage with provided data
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

t_statistic, degrees_of_freedom, p_value = paired_t_test(pre_intervention_scores, post_intervention_scores)
print("t-statistic:", t_statistic)
print("Degrees of freedom:", degrees_of_freedom)
print("p-value:", p_value)
````

 A political analyst wants to determine if there is a significant association between age groups and voter
preferences Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.


Use the below code to generate data:

```python

np.random.seed(0)

age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

````python
import numpy as np

np.random.seed(0)

# Generate age groups and voter preferences for 500 voters
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

````

````python
from scipy.stats import chi2_contingency

def chi_square_test(age_groups, voter_preferences):
    # Create a contingency table
    contingency_table = np.zeros((len(np.unique(age_groups)), len(np.unique(voter_preferences))))

    # Populate the contingency table
    for i, age_group in enumerate(np.unique(age_groups)):
        for j, preference in enumerate(np.unique(voter_preferences)):
            contingency_table[i, j] = np.sum((age_groups == age_group) & (voter_preferences == preference))

    # Perform the Chi-Square test
    chi2, p_value, dof, expected = chi2_contingency(contingency_table)

    # Interpret the results
    if p_value < 0.05:
        print("There is a significant association between age groups and voter preferences.")
    else:
        print("There is no significant association between age groups and voter preferences.")

    return chi2, p_value, dof, expected

# Example usage with generated data
chi2_value, p_value, degrees_of_freedom, expected_values = chi_square_test(age_groups, voter_preferences)
print("Chi-Square Value:", chi2_value)
print("Degrees of Freedom:", degrees_of_freedom)
print("p-value:", p_value)
print("Expected Values:")
print(expected_values)
````
