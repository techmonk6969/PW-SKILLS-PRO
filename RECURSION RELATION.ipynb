Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.
We will first expand the recurrence relation until we reach T(2):

T(2) = 3T(2-1) + 12*2
= 3T(1) + 24

Now, we continue expanding T(1) using the same formula:

T(1) = 3T(1-1) + 12*1
= 3T(0) + 12

Since T(0) = 5 (given), we substitute the value of T(0) into the equation:

T(1) = 3*5 + 12
= 15 + 12
= 27

Now, substitute the value of T(1) back into T(2) to get the final value:

T(2) = 3*27 + 24
= 81 + 24
= 105

So, T(2) = 105.
